## For Specifics Web Sites

[Arguments Documentation](https://google-images-download.readthedocs.io/en/latest/arguments.html)

Using --specific_site argument

According to the reference there is a list of reliable websites for catching the medical images

https://www.derm101.com/image-library/

http://www.dermatlas.net/ 

http://www.dermis.net/dermisroot/en/home/index.htm 

http://www.meddean.luc.edu/lumen/MedEd/medicine/dermatology/melton/atlas.htm 

http://www.dermatoweb.net/ 

http://www.atlasdermatologico.com.br/ 

http://www.hellenicdermatlas.com/en/?params=en 


```python
SITES_LIST = ["https://www.derm101.com/dermquest/", 
              "http://www.dermatlas.net/", 
              "http://www.dermis.net/dermisroot/en/home/index.htm",
              "http://www.meddean.luc.edu/lumen/MedEd/medicine/dermatology/melton/atlas.htm",
              "http://www.dermatoweb.net/, 
              "http://www.atlasdermatologico.com.br/",
              "http://www.hellenicdermatlas.com/en/?params=en"]

type(SITES_LIST)
```

For 12 Specific skin lessions 

* Actinic Keratosis

* Basal cell carcinoma

* Dermatofibroma

* Hemangioma

* Intraepithelial carcinoma (Bowens disease)

* Lentigo

* Malignant melanoma

* Melanocytic nevus

* Pyogenic granuloma

* Seborrheic keratosis

* Squamous cell carcinoma

* Wart

```python
# importing the library
from google_images_download import google_images_download 

for site in SITES_LIST:
    print(site)
    response = google_images_download.googleimagesdownload() 

    # creating list of arguments
    arguments = {
        "keywords": "Actinic Keratosis, \
                    Basal cell carcinoma, \
                    Dermatofibroma,\
                    Hemangioma, \
                    Intraepithelial carcinoma, \
                    Bowens disease, \
                    Lentigo, \
                    Malignant melanoma, \
                    Melanocytic nevus, \
                    Pyogenic granuloma, \
                    Seborrheic keratosis, \
                    Squamous cell carcinoma, \
                    Wart",
        
        "limiti": 1000,
        "specific_site": str(site),
        "print_urls":True
    }

    # passing the arguments to the function
    paths = response.download(arguments)  
    print(paths)
```
